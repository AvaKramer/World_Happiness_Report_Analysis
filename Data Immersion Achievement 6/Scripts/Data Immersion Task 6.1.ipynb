Data consistency checks consist of importing data sets, investigating accuracy of columns, checking for mixed-type data, checking for missing values, checking for duplicates, and exporting checked data.

```python
#Import libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Define the path

path = r'C:\Users\avamk\Happiness Report Analysis'
```

# 1. 2015 World Happiness Data

```python
#Import happiness report 2015 data set

df_2015 = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'WHR_2015.csv'))
```

```python
df_2015.head()
```

```python
df_2015.describe()
```

```python
#Check for mixed-type data

for col in df_2015.columns.tolist():
  weird = (df_2015[[col]].applymap(type) != df_2015[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_2015[weird]) > 0:
    print (col)
```

No mixed-type data found in the 2015 data set.

```python
#Check for missing values

df_2015.isnull().sum()
```

No missing values found in the 2015 data set.

```python
#Check for duplicates

df_dups_15 = df_2015[df_2015.duplicated()]
```

```python
df_dups_15
```

No duplicates found in the 2015 data set.

```python
#Export WHR_2015_Checked

df_2015.to_csv(os.path.join(path, '02 Data','Prepared Data', 'WHR_2015_checked.csv'))
```
